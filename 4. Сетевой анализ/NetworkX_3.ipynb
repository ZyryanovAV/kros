# Домашнее задание по библиотеке NetworkX

```python
import networkx as nx

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

## Задание 1 (1 балл)
Создайте граф и добавьте в него $5$ вершин с городами России, у каждой должен быть атрибут (характеристика) `time_diff`, означающий временную разницу с Москвой в часах.

```python
a = nx.Graph()
a.add_nodes_from(['Калининград', 'Астрахань', 'Челябинск', 'Омск', 'Новосибирск'])
a.add_node('Калининград', time_diff= 1)
a.add_node('Астрахань', time_diff= 1)
a.add_node('Челябинск', time_diff= 2)
a.add_node('Омск', time_diff= 3)
a.add_node('Новосибирск', time_diff= 4)
nx.draw_networkx(a, node_color='red')
plt.axis('off')

print('Graph nodes:', a.nodes())
```

## Задание 2 (1 балл)
Выберите какую-нибудь конкретную авиакомпанию (Аэрофлот, S7, Utair, etc).

В созданный граф добавьте рёбра между выбранными $5$ю городами, между которыми есть авиарейсы данной компании. У каждого ребра должны быть следующие атрибуты:
- расстояние между городами в километрах `distance`
- время перелёта в минутах `time`

При желании можно добавить в граф также несколько городов, являющихся пересадочными пунктами.

```python
a.add_edge('Астрахань', 'Челябинск', distance = 13540, time = 32000)
a.add_edge('Астрахань', 'Омск', distance = 2012, time = 540)
a.add_edge('Астрахань', 'Новосибирск', distance = 2607, time = 1140)
a.add_edge('Омск', 'Новосибирск', distance = 608, time = 1380)
a.add_edge('Омск', 'Калининград', distance = 3303, time = 1260)
a.add_edge('Новосибирск', 'Калининград', distance = 3858, time = 1585)
a.add_edge('Калининград', 'Челябинск', distance = 2576, time = 1115)
```

```python
plt.figure(figsize=(12, 12)) 
nx.draw_networkx(a, node_color='red', node_size=250)
plt.axis('off')

print('Города:', a.nodes())
```

## Задание 3 (1 балл)
Выберите любую пару городов и постройте между ними следующие кратчайшие пути:
- без взвешивания
- со взвешиванием по расстоянию
- со взвешиванием по времени

```python
d = nx.shortest_path(a, 'Челябинск', 'Новосибирск')
print(f'Без взвешивания :{d}')
f = nx.shortest_path(a, 'Челябинск', 'Новосибирск', weight = 'distance')
print(f'С взвешиванием по расстоянию: {f}')
g = nx.shortest_path(a, 'Челябинск', 'Новосибирск', weight = 'time')
print(f'Со взвешиванием по времени: {g}')
nx.draw_networkx(a, node_color='lightblue')
```

```python
print('Без взвешивания:')
print(nx.shortest_path(g, 'Moscow', 'Berlin'))
print('С взвешиванием по расстоянию')
print(nx.shortest_path(g, 'Moscow', 'Berlin', weight = 'distance'))
print('С взвешиванием по времени')
print(nx.shortest_path(g, 'Moscow', 'Berlin', weight = 'time'))
nx.draw_networkx(g, node_color='lightblue')
```
